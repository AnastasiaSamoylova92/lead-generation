```python
from pathlib import Path

import numpy as np
import pandas as pd

pd.set_option("display.max_columns", 180)
pd.set_option("display.width", 180)
pd.set_option("display.float_format", lambda x: f"{x:,.4f}")

RECO_DIR = Path("../data/product_recommendation")
CLUSTER_DIR = Path("../data/cohort_clustering")
PROCESSED_DIR = Path("../data/processed")
OUTPUT_DIR = Path("../data/dashboard_datasets")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print("Reco directory:", RECO_DIR.resolve())
print("Cluster directory:", CLUSTER_DIR.resolve())
print("Processed directory:", PROCESSED_DIR.resolve())
print("Output directory:", OUTPUT_DIR.resolve())
```

```python
prefix = "cluster_"
df_customers_clustering = pd.read_csv(CLUSTER_DIR / "df_clustering_output.csv").add_prefix(prefix)
df_customers_clustering["ID_Customer"] = df_customers_clustering["cluster_ID_Customer"]
df_customers_raw = pd.read_csv(PROCESSED_DIR / "dim_customer_clean.csv")
dim_customer = df_customers_raw.merge(df_customers_clustering, how="left")
dim_customer = dim_customer.drop(columns=["cluster_Sales", "cluster_Units", "cluster_order_lines"])
dim_customer
#df_customers_clustering
```

```python
df_reco_0 = pd.read_csv(RECO_DIR / "lead_generation_final.csv")
df_reco = df_reco_0.drop(columns=["cohort_ML", "cluster", "Sales", "sf_total_events", "activity_ratio_lifetime", 
                                  "months_since_last_purchase", "active_months", "avg_monthly_revenue", "cluster_ID_Customer"])

dim_customer_full = dim_customer.merge(df_reco, how="left")
dim_customer_full
```

```python
# save Outputs
outputs = {
    "dim_customer_full": dim_customer_full}

for name, df in outputs.items():
    df.to_csv(OUTPUT_DIR / f"{name}.csv", index=False)

print("Saved files:")
for path in sorted(OUTPUT_DIR.glob("*")):
    print("-", path.name)
```
